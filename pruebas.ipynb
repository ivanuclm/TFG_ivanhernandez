```python
import pickle
import pandas as pd
```

```python
def load_and_concatenate_list(filenames):
    data = []
    for filename in filenames:
        with open(filename, 'rb') as f:
            data += pickle.load(f)
    return data
```

```python
tracks = load_and_concatenate_list(['output/tmp/tracks_WIP_batch1.pkl', 'output/tmp/tracks_WIP_batch2.pkl', 'output/tmp/tracks_WIP_batch3.pkl', 'output/tmp/tracks_WIP_batch4.pkl', 'output/tmp/tracks_WIP_batch5.pkl', 'output/tmp/tracks_WIP_batch6.pkl', 'output/tmp/tracks_WIP_batch7.pkl', 'output/tmp/tracks_WIP_batch8.pkl'])
print(f"Hay {len(tracks)} canciones extraídas")

tracks 

# remove dupes on track_id
tracks = pd.DataFrame(tracks)
tracks = tracks.drop_duplicates(subset=['id'])
print(f"Hay {len(tracks)} canciones únicas")
```

```python
track_features = load_and_concatenate_list(['output/tmp/track_features_WIP_1.pkl', 'output/tmp/track_features_WIP_2.pkl', 'output/tmp/track_features_WIP_3.pkl'])
print(f"Hay {len(track_features)} features extraídas")

track_features = pd.DataFrame(track_features)
track_features = track_features.drop_duplicates(subset=['id'])

print(f"Hay {len(track_features)} features únicas")

# check how many rows in track_features do not have features or they are empty
print(f"Hay {len(track_features[track_features['acousticness'].isnull()])} filas sin features")
print(f"Hay {len(track_features[track_features['acousticness'] == 0])} filas con features vacíos")
```

### TOP_ARTIST_IDS

```python
import openpyxl
from openpyxl import load_workbook

wb = load_workbook(filename = 'input/TOP_ARTISTAS_kworbnet_07052024.xlsm', data_only=True)
ws = wb['Hoja1']

artistas = []

for row in ws.iter_rows(min_row=2, max_row=ws.max_row, min_col=7, max_col=7):
    for cell in row:
        artistas.append(cell.value)

artistas

with open('input/artistas.pkl', 'wb') as f:
    pickle.dump(artistas, f)

df = pd.DataFrame(artistas, columns=['Artista'])
df.to_csv('input/artistas.csv', index=False)
```
